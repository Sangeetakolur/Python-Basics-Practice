# match statement

```python
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "YES"
        case 404:
            return "not found"
        case 500:
            return "internal server error"
        case _:
            return "unknown status"
print(http_status(500))
print(http_status(1000))

a = 5
b = 4
match a + b:
    case 9:
        print("sum is 9")
    case 10:
        print(" sum is not correct")
    case _:
        print("not found")

x = 8
y = 4
operator = "*"

match operator:
    case "+":
        print(x + y)
    case "-":
        print(x - y)
    case "*":
        print(x * y)
    case "/":
        print(x / y)
    case _:
        print("Invalid operator")

def http_status(code):
    match code // 100:
        case 2:
            return "OK"
        case 4:
            return "Client Error"
        case 5:
            return "Internal Server Error"
        case _:
            return "Unknown status"

print(http_status(500))
print(http_status(1000))



        
```
